# 3.1 c)
#### What is the difference between the three values of current computed at each node or line?

-The initial symmetrical short-circuit is the effective value of the short circuite, when the short circuit arises. At the point where the short circuit impedance is equal to the value at time zero. 

-The short-circuit current peak is the largest possible momentary value of the short circuit.

-The equivalent thermal short-circuit current is the constant value that depends on the thermal effect of the DC component and the fault duration.

# 3.1 d)
#### Which standard is used to compute the short-circuit currents? Briefly explain the methodology based on the documentation, the slides of the course, and material you find online.

The standars which is used to compute the short-circuit currents is called IEC 60909. The methode to compute the short-circuit current is called superposition method. It is an exact method fot the calculation and consists of three steps. 
In the first step the currents, voltages and internal voltages for steady-state are calculated. This consider the impedances, power supply feeders and node loads.
In the second step the voltage and current at the fault location befor the fault are determined with a negative sign. This voltage is the only voltage source in the network and the internal voltages are short-circuited. 
In the third step both conditions are superimposed, resulting in a zero voltage at the fault location. The superposition of the currents also results in zero.

The disadvantage of the superposition method is, that the voltage ratios and the loading condition of the network must be known for the steady-state before the occurrence of the short circuit. And the date for effective and reactive power, node voltages and steps settings of the transformer are often difficult to determine.
